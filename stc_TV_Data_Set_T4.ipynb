```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# --- 1. الهدف: تحديد اتجاه تغير البيانات بعد إطلاق ميزة جديدة ---
# لنفترض أننا أطلقنا "وضع التركيز" لمساعدة المستخدمين على تقليل وقت استخدام التطبيق.
# هدفنا هو أن نرى ما إذا كانت الميزة فعالة.

# إنشاء بيانات وهمية لتمثيل القصة
dates = pd.date_range(start='2023-05-01', end='2023-05-30')
launch_date = '2023-05-15'

# قبل الإطلاق: وقت استخدام مرتفع (متوسط 220 دقيقة)
before_launch_time = np.random.randint(200, 240, size=14)
# بعد الإطلاق: وقت استخدام منخفض (متوسط 140 دقيقة)
after_launch_time = np.random.randint(120, 160, size=16)

screen_time = np.concatenate([before_launch_time, after_launch_time])
df = pd.DataFrame({'date': dates, 'screen_time_minutes': screen_time})
```

# --- بدء السرد القصصي (ما سيتم عرضه في العرض التقديمي) ---

```python
# --- 2 & 3. تحديد الجمهور والشخصيات والمشكلة ---
print("--- قصة بيانات: تأثير ميزة 'وضع التركيز' ---")
print("\nالجمهور المستهدف: فريق إدارة المنتج.")
print("الشخصية الرئيسية: مستخدمو تطبيقنا الذين كانوا يقضون وقتاً طويلاً عليه.")
print("المشكلة: كان متوسط وقت الاستخدام اليومي مرتفعاً، مما قد يسبب إرهاقاً للمستخدمين.")
```

```python
# --- 4. الالتزام بالخط الزمني وعرض النتائج ---
print(f"\nالحل: في تاريخ {launch_date}، أطلقنا ميزة 'وضع التركيز'.")
print("الآن، دعونا نرى ما الذي حدث بعد ذلك...")

# حساب النتائج
avg_before = df[df['date'] < launch_date]['screen_time_minutes'].mean()
avg_after = df[df['date'] >= launch_date]['screen_time_minutes'].mean()
percentage_drop = ((avg_before - avg_after) / avg_before) * 100

print(f"\nالنتائج:")
print(f"- متوسط وقت الاستخدام قبل الإطلاق: {avg_before:.0f} دقيقة/يوم.")
print(f"- متوسط وقت الاستخدام بعد الإطلاق: {avg_after:.0f} دقيقة/يوم.")
print(f"==> هذا يمثل انخفاضاً بنسبة {percentage_drop:.1f}%! الهدف تحقق.")
```

```python
# --- 5. استخدام العناصر المرئية البسيطة ---
print("\nوهذا هو شكل القصة في رسم بياني:")
plt.style.use('seaborn-v0_8-whitegrid')
fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(df['date'], df['screen_time_minutes'], marker='o', linestyle='-', label='وقت الاستخدام اليومي')
# إضافة خط عمودي يوضح تاريخ إطلاق الميزة (عنصر قصصي قوي جداً)
ax.axvline(pd.to_datetime(launch_date), color='red', linestyle='--', label=f'إطلاق وضع التركيز ({launch_date})')

ax.set_title('تأثير "وضع التركيز" على وقت استخدام التطبيق')
ax.set_xlabel('التاريخ')
ax.set_ylabel('وقت الاستخدام (بالدقائق)')
ax.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
